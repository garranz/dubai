```python
# ======= Parameters =============
z0 = 0.0002   # Surface roughness length
u_star = 0.5  # Friction velocity
k = 0.4       # von Karman constant
#=================================
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def wind_profile(z_, z0_, u_star_, k_):
    """
    Computes the wind speed at height z in the atmospheric boundary layer
    using the logarithmic wind profile.

    Parameters:
    - z : height above ground (m)
    - z0 : roughness length (m)
    - u_star : friction velocity (m/s)
    - k : von Karman constant (~0.4)

    Returns:
    - u : wind speed at height z (m/s)
    """
    return (u_star_ / k_) * np.log(z_ / z0_)
```

```python
def print_abl_profile():
    heights = np.linspace(1, 1000, 40)  # heights from 1 m to 1000 m
    print(f"{'Height (m)':>12} | {'Wind Speed (m/s)':>16}")
    print("-" * 32)
    u = np.zeros_like(heights)
    for i, z in enumerate(heights):
        if z > 0:
            u[i] = wind_profile(z, z0, u_star, k)
            print(f"{z:12.1f} | {u[i]:16.2f}")
    return heights, u
```

```python
def plot_abl_profile(heights, wind_speeds):
    plt.figure(figsize=(6,8))
    plt.plot(wind_speeds, heights, marker='o', color='blue')
    plt.xlabel("Wind Speed (m/s)")
    plt.ylabel("Height (m)")
    plt.title("Atmospheric Boundary Layer Wind Profile")
    plt.grid(True)
    plt.gca().invert_yaxis()
    plt.gca().invert_yaxis()
    plt.show()
```

```python
print("Atmospheric Boundary Layer Wind Profile")
print("---------------------------------------")
h,u = print_abl_profile()
plot_abl_profile(h,u)
```
